#### Create New Columns

References 

- Add Columns: https://pandas.pydata.org/docs/getting_started/intro_tutorials/05_add_columns.html
- Drop Rows: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html

```python
#Drop cancelled and diverted flights from the dataframe since the Avg Speed will calculated as function of AirTime.
airline_df.drop(airline_df[airline_df.Cancelled == 1].index, inplace=True)
airline_df.drop(airline_df[airline_df.Diverted == 1].index, inplace=True)
```

```python
#Check number of rows in dataframe after dropping cancelled and diverted flights.
airline_df.shape
```

I confirmed that the units of Distance are miles by checking the first row's origin and destination using an online calculator: https://www.airmilescalculator.com/distance/iad-to-tpa/

The AirTime units are minutes, so this must be divided by 60 to provide the speed units in miles per hour (mph).


```python
#Add average flight speed in miles per hour.
airline_df["AvgFlightSpeed"] = (
    airline_df["Distance"] / (airline_df["AirTime"]/60)
)

airline_df.head()
```

```python
airline_df.AvgFlightSpeed.describe().round(2)
```

There are unexpected infinity values in the average flight speed, so we must examine if there are flights that do not have an AirTime recorded.

```python
#Check summary statistics of 
airline_df.AirTime.describe().round(2)
```

```python
#Check which rows have no AirTime recorded.
airline_df[airline_df.AirTime == 0]
```

There appears to be a problem with the data collection of the AirTime for 7 flights. I will replace the infinity values with NAs.

```python
#Replace infinity values in dataset with NA
airline_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#Check Average Flight Speed summary statistics after replacing the infinity values.
airline_df.AvgFlightSpeed.describe().round(2)
```

Even after removing infinity values, there are still some values for average flight speeds that don't make sense. Outliers probably exist within the data. Let's take a look at some of the top values for average flight speed.

In this case, the AirTime is still an issue because there are values which are very low and don't make sense. The AvgFlightSpeeds are impossible, because the fastest commerical planes travel at top speed of just over 650 miles per hour.

The minimum value in the DepDelay column is 6 minutes, which means 100% of the obeservations in the dataset experience delayed flights upon departure. 

```python
#Add whether or not flight is delayed at departure.
airline_df["DepartureDelay"] = (
    airline_df["DepDelay"] > 0.0
)

airline_df.head()
```

```python
#Summary of Departure Delay
#convert Trues & Falses to 1s and 0s.
airline_df["DepartureDelay"] = airline_df["DepartureDelay"].astype(int)
num_dep_delay = sum(airline_df.DepartureDelay)
print ("There are", num_dep_delay, "flights delayed at departure in the dataset out of",\
       airline_df.shape[0], "observations.")
```

```python
airline_df.ArrDelay.describe().round()
```

```python
#Add whether or not flight is delayed at arrival.
airline_df["ArrivalDelay"] = (
    airline_df["ArrDelay"] > 0.0
)

airline_df.head()
```

```python
#Summary of Arrival Delay
#convert Trues & Falses to 1s and 0s.
airline_df["ArrivalDelay"] = airline_df["ArrivalDelay"].astype(int)
num_arr_delay = sum(airline_df.ArrivalDelay)
print ("There are", num_arr_delay, "flights delayed at arrival in the dataset out of",\
       airline_df.shape[0], "observations.")
print ((num_arr_delay/airline_df.shape[0]*100), "% of flights experience arrival delays")
```

#### Table of Airlines with Most Accumulated Delays

Since 100% of flights were delayed at departure, I will evalute arrival delays by airline carrier. Since each carrier operates a different number of flights per year, I will sort the carriers by worst performance as a percentage. I will evaluate the ratio of the number of flights delayed at arrival to the number of annual flights.

```python
grouped = airline_df.groupby('UniqueCarrier')
carrier_delays = (grouped['ArrivalDelay'].agg(np.sum))
carriers_df["annual delays"] = carrier_delays
carriers_df["percent of flights delayed"] = (carriers_df["annual delays"]/carriers_df["annual flights"])*100
carrier_delays = carriers_df.sort_values(by=['percent of flights delayed'], ascending=False).copy()
carrier_delays
```

The worst 3 performers for arrival delays based on the percentage of all their flights are HA: Hawaiian Airlines, YV: Mesa Airlines, and OH: PSA Airlines.

The top 3 perofmers for arrival delays are CO: North-Western Cargo International Airlines, US: Puerto Rico International Airlines, and WN: Southwest Airlines.

#### Final Questions

```python
#What are the longest flights?
grouped = airline_df.groupby(['UniqueCarrier','FlightNum','Origin','Dest'])
longest_flights = (grouped['AirTime'].agg(np.mean))
longest_flights = longest_flights.sort_values(ascending=False)
longest_flights.head(10)
```

I chose to evalute by mean air time between destinations for a given airline and flight number. The longest flights by travel time are between the mainland United States and Hawaii. HNL is in Honolulu and OGG is on the island of Maui. 

It will also be interesting to look at the longest flights according to distance between destinations, regardless of the airline or the flight number.

```python
grouped = airline_df.groupby(['Origin','Dest'])
longest_flights = (grouped['Distance'].agg(np.mean))
longest_flights = longest_flights.sort_values(ascending=False)
longest_flights.head(10)
```

It is not surprising that the longest flights by distance are between the east coast or midwest of the mainland United States and Hawaii. The three longest flights are between Newark, NJ and Honolulu, Atlanta and Honololulu, and Chicago and Honolulu.

```python
#What are the most delayed flights?
grouped = airline_df.groupby(['UniqueCarrier','FlightNum','Origin','Dest'])
delayed_flights = (grouped['ArrDelay'].agg(np.mean))
delayed_flights = delayed_flights.sort_values(ascending=False)
delayed_flights.head(10)
```

The table above shows the worst 10 average arrival delays in minutes for a given carrier and flight number for an origin and destination in descending order. .
